# Chicago Crime Stats

```python
# Define the database URL
db_url = "postgresql://postgres:admin@localhost:5432/db_project"
# Create the engine object
engine = create_engine(db_url)
```

```python
sql = """select block, primary_type, description, location_description, years, latitude, longitude from crime_data
where year between'2017' and '2020'"""
df3 = pd.read_sql(sql, engine)

df3

df3.dropna()

#Converting years, latitude, longitude into ints and floats in DataFrame
convert_dict = {'years': int, 'latitude': float, 'longitude': float, 'primary_type': str}

df3 = df3.astype(convert_dict)
print(df3.dtypes)

df3
```

```python
df3.hvplot.points(
    'longitude',
    'latitude',
    geo=True,
    tiles='OSM',
    hover_cols='all',
    frame_width=700,
    frame_height=500,
    title='Intecactive Map Plotting Chicago',
    groupby='block'
)
```
